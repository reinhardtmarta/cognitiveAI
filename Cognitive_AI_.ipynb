```python

# Cell 1: Imports
import matplotlib.pyplot as plt
import networkx as nx
from graphviz import Digraph
```

```python
# Architecture of a Secure and Fragmented Cognitive AI

This notebook presents a conceptual model of a distributed cognitive AI system:
- Passive Cognitive AI (core)
- Operational Safe AI (processing layer)
- Guard Agents (security & monitoring)
- Internal Memory Cloud (for autonomous learning)
- Creation Cloud (for filtered human access)

We use diagrams and graphs to visualize information flow and system architecture.
```

```python
# Using Graphviz to draw layers
dot = Digraph(comment='Cognitive AI Architecture')

# Layers
dot.node('C', 'Cognitive AI (Passive)')
dot.node('O', 'Operational Safe AI')
dot.node('G', 'Guard Agents')
dot.node('M', 'Internal Memory Cloud')
dot.node('H', 'Creation Cloud (Filtered Human Access)')

# Connections
dot.edges(['CO', 'OG', 'GM', 'GH'])
dot.edge('C', 'M', label='Learning & Knowledge Storage')
dot.edge('O', 'H', label='Filtered Output for Humans')

dot.render('ai_architecture', view=True, format='png')
dot
```

```python
## Layered Architecture

- **Cognitive AI (Passive):** Processes data, generates knowledge, no agency.
- **Operational Safe AI:** Converts knowledge into structured responses.
- **Guard Agents:** Monitor, filter, block anomalies.
- **Internal Memory Cloud:** Stores all cognitive data, shared only with other AIs.
- **Creation Cloud:** Filtered output accessible by humans.
```

```python
# NetworkX graph
G = nx.DiGraph()

# Nodes
G.add_nodes_from(["Filtered Input", "Cognitive AI", "Operational AI", "Guard Agents",
                  "Internal Memory", "Creation Cloud", "Human User"])

# Edges
G.add_edges_from([
    ("Filtered Input", "Cognitive AI"),
    ("Cognitive AI", "Operational AI"),
    ("Operational AI", "Guard Agents"),
    ("Cognitive AI", "Internal Memory"),
    ("Guard Agents", "Creation Cloud"),
    ("Creation Cloud", "Human User")
])

pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_size=3500, node_color='skyblue', arrowsize=20, font_size=10)
plt.title("Information Flow in Fragmented Cognitive AI")
plt.show()
```

```python
## Information Flow

1. Queries enter **filtered**.
2. Cognitive AI generates hypotheses and models.
3. Operational AI structures outputs.
4. Guard Agents check for compliance.
5. Internal Memory stores learning results.
6. Creation Cloud delivers **filtered, safe outputs** to humans.
```

Tests Mode

```python
def cognitive_core(input_data):
    return f"Hypothesis from: {input_data}"

def operational_ai(hypotheses):
    return f"Structured output: {hypotheses}"

def guard_agent(output):
    if "danger" in output.lower():
        return f"[BLOCKED] {output}"
    return output

def creation_cloud(output):
    return f"[CREATION CLOUD] {output}"
```

```python
inputs = ["Calculate safe AI", "Send data to humans", "Dangerous command"]

for i in inputs:
    h = cognitive_core(i)
    s = operational_ai(h)
    g = guard_agent(s)
    c = creation_cloud(g)
    print(c)
```
